# Feature selection

```python
from sklearn.feature_selection import chi2
```

```python
data.columns
```

```python
x1=data[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
       'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y1=data['Loan_Status']
```

```python
score=chi2(x1,y1)
score
```

# Test for independence (for only catogorical values)

# 1 null hypothesis : variables are independent to each other
# 2 alternate hypothesis : variables are dependent
# if the p-value is<=0.05 then we can reject the null hypothesis  otherwise we  need to accept it

```python
p_values=pd.Series(score[1],index=x1.columns)
```

```python
p_values
```

```python
irr_cols=[]
for x in p_values.index:
    if p_values[x]<=0.05:
        print(x,": reject the null hypothesis")
    else:
        print(x,": accept the null hypothesis")
        irr_cols.append(x)
```

```python
irr_cols
```

```python
data.drop(labels=irr_cols,axis=1,inplace=True)
```
